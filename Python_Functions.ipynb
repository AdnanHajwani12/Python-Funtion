# **Practical Ques**

```python
#1. Sum of all even numbers in a list
def sum_even(numbers):
    total = 0
    for n in numbers:
        if n % 2 == 0:
            total += n
    return total

print(sum_even([1, 2, 3, 4, 6]))
```

```python
#2. Reverse a string
def reverse_string(s):
    return s[::-1]

print(reverse_string("python"))
```

```python
#3. Squares of each number in a list
def square_list(nums):
    result = []
    for n in nums:
        result.append(n * n)
    return result

print(square_list([1, 2, 3, 4]))
```

```python
#4. Check prime number (1 to 200)
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

print(is_prime(17))
```

```python
#5. Fibonacci iterator class
class Fibonacci:
    def __init__(self, terms):
        self.a = 0
        self.b = 1
        self.count = 0
        self.terms = terms

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

for num in Fibonacci(5):
    print(num)
```

```python
#6. Generator for powers of 2
def powers_of_two(exp):
    for i in range(exp + 1):
        yield 2 ** i

for p in powers_of_two(5):
    print(p)
```

```python
#7. Generator that yields each line of a string
def read_lines(text):
    for line in text.split("\n"):
        yield line

data = "Hello\nWelcome\nPython"
for l in read_lines(data):
    print(l)
```

```python
#8. Lambda to sort list of tuples by second element
data = [(1, 3), (4, 1), (2, 2)]
sorted_data = sorted(data, key=lambda x: x[1])
print(sorted_data)
```

```python
#9. Convert Celsius to Fahrenheit using map()
celsius = [0, 10, 20, 30]
fahrenheit = list(map(lambda c: (c * 9/5) + 32, celsius))
print(fahrenheit)
```

```python
#10. Remove vowels using filter()
text = "python programming"
result = ''.join(filter(lambda x: x.lower() not in 'aeiou', text))
print(result)
```

```python
#11. Book shop accounting (lambda + map)
orders = [
    (34587, 4, 40.95),
    (98762, 5, 56.80),
    (77226, 3, 32.95),
    (88112, 3, 24.99)
]

result = list(map(
    lambda x: (x[0], x[1]*x[2] if x[1]*x[2] >= 100 else x[1]*x[2] + 10),
    orders
))

print(result)
```
